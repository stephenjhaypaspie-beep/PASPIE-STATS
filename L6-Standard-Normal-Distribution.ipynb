```python
#import and packages

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

## Z-Score

```python
#creating a dataframe
df = pd.DataFrame([1,2,2,3,3,3,4,4,5],
                  columns = ['Dataset'])
df
```

```python
#Summary of Statistics
df.describe()
```

```python
#Z-Score
df['Z-Score'] = stats.zscore(df['Dataset'],ddof=1)
df
```

```python
df['Z-Score'].std(ddof=1)
```

```python
#load Dataset
df = pd.read_csv(r'csv/current-test.csv',
                 delimiter = ',',
                 index_col='Sample')
df
```

```python
df['Current'][2]
```

```python
df.describe()
```

```python
# Z-Score
df['Z-Score'] = stats.zscore(df['Current'],ddof=1)
df
```

```python
# Z-Score summary of statistics
df['Z-Score'].describe()
```

## Histogram

```python
#Histogram Plot
plt.hist(df['Current'], bins=5,
         alpha = 0.8,
         color = "red",
         edgecolor = "black")

plt.title('Histogram Of Current Values',
          font= 'Palatino Linotype',
          fontsize = 12)
plt.xlabel('Current (mA)')
plt.ylabel('Frequency')
plt.show()
```

## Normal Distribution

```python
#normal distribution curve
mean = df['Current'].mean()
std = df['Current'].std()

#X-axis
x_min = df['Current'].min()
y_max = df['Current'].max()

x = np.linspace(x_min, y_max, 100)
p = stats.norm.pdf(x, mean, std)
plt.plot(x,p,
         color = "Green",
         linewidth = 2)
plt.title('Normal Distribution Of Current Value',
          font= 'Palatino Linotype',
          fontsize = 12)
plt.xlabel('Current (mA)')
plt.ylabel('Frequency')
plt.show()
         
```

```python
#histogram plot 
plt.hist(df['Current'],bins = 5,
         density = True,
         alpha = 0.8,
         color= "Green",
         edgecolor = "Black")

plt.title('Normal Distribution Of Current Value',
          font= 'Palatino Linotype',
          fontsize = 12)
plt.xlabel('Current (mA)')
plt.ylabel('Frequency')

p = stats.norm.pdf(x, mean, std)
plt.plot(x,p,
         color = "blue",
         linewidth = 2)

plt.show()
```

## Standard Normal Distribution N(0,1)

```python
#Standard Normal Distribution N(0,1)
#x-axis
x_max = df['Z-Score'].max()
x_min = df['Z-Score'].min()
x = np.linspace(x_min, x_max, 100)

p = stats.norm.pdf(x, df['Z-Score'].mean(),df['Z-Score'].std(ddof=1))
plt.plot(x,p,
         color = "blue",
         linewidth = 2)

plt.title('Standard Normal Distribution N(0,1)',
          font= 'Palatino Linotype',
          fontsize = 12)
plt.xlabel('Z-Score')
plt.ylabel('Probability Density')
plt.show()
```
