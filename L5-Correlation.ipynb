```python
#import and packages

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
#load dataset
df = pd.read_csv(r'csv/current-power.csv' , delimiter= ",")

df
```

```python
#summary of statistics 
df.describe()
```

```python
df.cov()
```

```python
x = df['Current']
y = df['Power']

plt.figure()
plt.scatter(x,y)
plt.title("Device Current vs Power Response")
plt.ylabel('Power (W)')
plt.xlabel('Current (A)')
plt.show()
```

```python
#correlation coefficient of current and power

df.corr()
```

```python
#Regression Line 
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
line = slope * x + intercept

#Scatter Plot 
plt.figure(figsize=(7,4))
plt.scatter(x,y,
            color = 'Red',
            alpha = 0.6,
            label = 'Data Points')

#line Plot
plt.plot(x, line,
         color = 'Blue',
         label=f'Regression Line (r={r_value:.3f})')
        

plt.title('Device Current vs Power Response')
plt.ylabel('Power (W)')
plt.xlabel('Current (A)')
plt.legend()
plt.show()
```

## Air Quality

```python
# Load Dataset
air = pd.read_csv(r"csv\air-quality.csv",
                 delimiter=",")
air
```

## NOx - Humidity

```python
#Regression Line
x = air ['NOx']
y = air ['Relative_Humidity']

slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
line = slope * x + intercept

#Scatter Plot 
plt.figure(figsize=(7,4))
plt.scatter(x,y,
            color = 'Yellow',
            alpha = 0.2,
            label = 'Data Points')

#line Plot
plt.plot(x, line,
         color = 'Blue',
         label=f'Regression Line (r={r_value:.3f})')
        

plt.title('NOx vs Humidity Correlation')
plt.ylabel('Relative Humidity (%)')
plt.xlabel('NOx  (ug/cu.m)')
plt.legend()
```

## NO2 - Humidity

```python
x = air ['NO2']
y = air ['Relative_Humidity']

slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
line = slope * x + intercept

#Scatter Plot 
plt.figure(figsize=(7,4))
plt.scatter(x,y,
            color = 'Orange',
            alpha = 0.2,
            label = 'Data Points')

#line Plot
plt.plot(x, line,
         color = 'Blue',
         label=f'Regression Line (r={r_value:.3f})')
        

plt.title('NO2 vs Humidity Correlation')
plt.ylabel('Relative Humidity (%)')
plt.xlabel('NO2  (ug/cu.m)')
plt.legend()
```
